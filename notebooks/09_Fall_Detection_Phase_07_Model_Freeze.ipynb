```python
# ==========================================
# PHASE 07: MODEL FREEZE (MANDATORY)
# ==========================================
# Goal: Lock the trained models (v1.0) so they are
# safe for the final Web Demo (Phase 8).
# ==========================================

import os
import shutil
from google.colab import drive

# 1. MOUNT DRIVE
drive.mount('/content/drive')

# 2. PATHS
BASE_PATH = "/content/drive/MyDrive/Human-Fall-Recognition"
SOURCE_MODELS = os.path.join(BASE_PATH, "Models")
DEPLOY_DIR    = os.path.join(BASE_PATH, "Deployment_Frozen")

os.makedirs(DEPLOY_DIR, exist_ok=True)

print("🚀 PHASE 07: FREEZING MODELS FOR THESIS DEFENSE...")

# 3. IDENTIFY MODELS TO FREEZE
# We look for the files generated in Phase 4
targets = [
    "model_baseline.keras",
    "model_enhanced.keras",
    "yolov8n-pose.pt" # We keep a copy of the specific YOLO version too
]

# 4. EXECUTE FREEZE (COPY)
for file_name in targets:
    src = os.path.join(SOURCE_MODELS, file_name)
    dst = os.path.join(DEPLOY_DIR, file_name)

    # Check if source exists in Models folder
    if os.path.exists(src):
        shutil.copy2(src, dst)
        print(f"✅ Frozen: {file_name}")
        print(f"   Source: {src}")
        print(f"   Dest  : {dst}")
    else:
        # If YOLO isn't in Models folder, it might be in the current working directory
        if "yolo" in file_name:
            if os.path.exists(file_name): # Local runtime check
                shutil.copy2(file_name, dst)
                print(f"✅ Frozen: {file_name} (from local runtime)")
            else:
                 # Last resort: check if it's in the base path
                 root_src = os.path.join(BASE_PATH, file_name)
                 if os.path.exists(root_src):
                     shutil.copy2(root_src, dst)
                     print(f"✅ Frozen: {file_name} (from Project Root)")
                 else:
                     print(f"⚠️ Warning: {file_name} not found to freeze. (You can download it again for Phase 8)")
        else:
            print(f"❌ ERROR: Critical model missing: {file_name}")

# 5. CREATE 'VERSION_INFO.txt'
# This proves to the examiner that these models are the final ones.
info_path = os.path.join(DEPLOY_DIR, "VERSION_INFO.txt")
with open(info_path, "w") as f:
    f.write("PROJECT: Human Fall Recognition (Skeleton+LSTM)\n")
    f.write("STATUS: FROZEN FOR THESIS DEFENSE\n")
    f.write("VERSION: v1.0\n")
    f.write("DATE: 2026-01-30\n")
    f.write("CONTENTS:\n")
    f.write("- model_baseline.keras: Standard LSTM\n")
    f.write("- model_enhanced.keras: Bi-LSTM (Final Choice)\n")
    f.write("- yolov8n-pose.pt: Pose Estimation Engine\n")

print(f"📝 Version Info Saved: {info_path}")

print("\n" + "="*40)
print("❄️  MODELS SUCCESSFULLY FROZEN  ❄️")
print("="*40)
print(f"files are safe in: {DEPLOY_DIR}")
print("You are now ready for PHASE 08 (Web App Demo).")
```
